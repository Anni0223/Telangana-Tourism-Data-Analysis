```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('retina', quality=100)
```

```python
district_details = pd.read_excel('Demographics and facilities raw.xlsx')
```

```python
district_details
```

```python
x=district_details['2019 DV']
y=district_details['Cagr']
n=district_details['District Acronym']
```

```python
from matplotlib.ticker import FuncFormatter
import matplotlib.pyplot as plt
def millions(x, pos):
    'The two args are the value and tick position'
    return '%1.0fM' % (x * 1e-6)


formatter = FuncFormatter(millions)
```

```python
fig, ax = plt.subplots(figsize=(14,9))
ax.scatter(x, y)
fig.patch.set_facecolor('#F5F5F5')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.xaxis.set_major_formatter(formatter)

ax.set_xlabel('Domestic Visitors 2019', labelpad=15, color='#333333')
ax.set_ylabel('CAGR of Domestic Visitors', labelpad=15, color='#333333')

for i, txt in enumerate(n):
    ax.annotate(txt, (x[i], y[i]))
```
