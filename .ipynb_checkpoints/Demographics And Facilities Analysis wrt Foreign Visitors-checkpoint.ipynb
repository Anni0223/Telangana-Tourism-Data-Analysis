```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('retina', quality=100)
```

```python
district_details = pd.read_excel('Demographics and facilities for FV.xlsx')
```

```python
district_details
```

```python
x=district_details[2019]
y=district_details['Cagr']
n=district_details['District Acronym']
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.scatter(x, y)
ax.plot([-0.285236,-0.285236,-0.285236,-0.285236,-0.285236,-0.285236,-0.285236])
ax.set_xlabel('Domestic Visitors', labelpad=15, color='#333333')
ax.set_ylabel('CAGR of Domestic Visitors', labelpad=15, color='#333333')

for i, txt in enumerate(n):
    ax.annotate(txt, (x[i], y[i]))
```
